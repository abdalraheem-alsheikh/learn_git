```python
def line_equation(r1,x2,y1,y2):
  m=(y2-y1)/(x2-r1)
  b=y1-m*r1
  e=f"y={m}x+{b}"
  return e
```

```python
def calculate_grade(degrees):
  grades=[]
  for i in degrees:
    if i>=90:
      grades.append("A")
    elif i>=80:
      grades.append("B")
    elif i>=70:
      grades.append("C")
    else:
      grades.append("F")
  print(grades)
```
